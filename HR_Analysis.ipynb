```python
def salary_group(salary): #func to group salary
    if salary <= 53000:
        return "Low"
    elif salary <= 86000:
        return "Mid"
    else:
        return "High"

df['Salary_Group'] = df['Salary'].apply(salary_group) #running the func on Salary column in dataframe
df.groupby('Salary_Group')['Employee_ID'].count().sort_values(ascending=False) #counting employees in each salary group
```

```python
def exp_group(exp): #func to create experience ranges
    if exp < 2:
        return "0-2"
    elif exp <= 5:
        return "3-5"
    elif exp <= 10:
        return "6-10"
    else:
        return "10-14"

df['Experience_Group'] = df['Experience (Years)'].apply(exp_group) #running the function on the Experience (Years) column in the dataset
df['Experience_Group'].value_counts().sort_values(ascending=False) #counting employees in each exp years range
```
